```python
import numpy as np
import pandas as pd
```

## 오징어 두바이유, 환율, P_PRICE 상관관계

```python
squid = pd.read_csv('process_codes/오징어 주별 평균단가.csv', encoding='cp949')
data = pd.read_csv('data/raw_data/외부데이터모음(2015~2021).csv', encoding='cp949')

print(squid)
print(data)
```

```python
idx = data[(data['REG_DATE'] == '2016-02-08') | (data['REG_DATE'] == '2017-10-02')].index
print(idx)

idx_1 = data[(data['REG_DATE'] >= '2021-01-04')].index
print(idx_1)
```

```python
data.drop(idx, inplace=True)
data.drop(idx_1, inplace=True)

data.reset_index(drop=True, inplace=True)

print(squid.shape)
print(data.shape)
```

```python
x_squid = data['원달러환율']
y_squid = data['두바이유']
z_squid = squid['P_PRICE']
```

```python
print(np.corrcoef(x_squid, z_squid)[0, 1])

print(np.corrcoef(y_squid, z_squid)[0, 1])
```
