```python
import pandas as pd
```

```python
df = pd.read_csv('RAW_DATA/흰다리새우_데이터_최종셋.csv')
df = df.drop('Unnamed: 0', axis = 1)
df['REG_DATE'] = pd.to_datetime(df['REG_DATE'], format="%Y-%m-%d")
```

## 주별 평균단가 

```python
date_price_mean  = df.groupby('REG_DATE').mean()

date_price_mean.to_csv('RAW_DATA/변수/흰다리새우_주별_평균단가.csv')
```

## 국가별 평균단가 및 카운트

### 평균단가

```python

nation_mean = df.groupby(['REG_DATE', 'CTRY_2'])['P_PRICE'].mean()

n_mean = nation_mean.unstack()

n_mean.to_csv('RAW_DATA/변수/흰다리새우_국가별_평균.csv')
```

### 카운트

```python
nation_count = df.groupby(['REG_DATE', 'CTRY_2'])['P_PRICE'].count()

n_count = nation_count.unstack()

for i in n_count.columns:
    n_count.rename(columns= {i: i+'C'}, inplace=True)


print(n_count)
n_count.to_csv('RAW_DATA/변수/흰다리새우_국가별_카운트.csv')
```

## 주별 국가 타입별 평균단가 및 카운트

### 평균

```python
nation_type_mean = df.groupby(['REG_DATE', 'CTRY_2', 'P_IMPORT_TYPE'])['P_PRICE'].mean()

n_t_mean = nation_type_mean.unstack()

n_t_mean_1 = n_t_mean.unstack()

column_list = n_t_mean_1.columns

column_list_1 = []
for i in column_list:
    x = i[0] + '_' + i[1]
    column_list_1.append(x)

column_list_1

n_t_mean_1.columns = column_list_1
print(n_t_mean_1.shape)

n_t_mean_2 = n_t_mean_1.transpose()

print(n_t_mean_2)

n_t_mean_2.dropna(axis=0, how="all", inplace= True)
n_t_mean_2 = n_t_mean_2.transpose()

print(n_t_mean_2.shape)


n_t_mean_1.to_csv('RAW_DATA/변수/흰다리새우_국가수입형태별_평균.csv')
```

### 카운트

```python
nation_type_count = df.groupby(['REG_DATE', 'CTRY_2', 'P_IMPORT_TYPE'])['P_PRICE'].count()

n_t_count = nation_type_count.unstack()

n_t_count_1 = n_t_count.unstack()

column_list = n_t_count_1.columns

column_list_1 = []
for i in column_list:
    x = i[0] + '_' + i[1] + 'C'
    column_list_1.append(x)

column_list_1

n_t_count_1.columns = column_list_1

print(n_t_count_1)

n_t_mean_1.to_csv('RAW_DATA/변수/흰다리새우_국가수입형태별_카운트.csv')
```

## 환율 데이터

```python
df = pd.read_csv('RAW_DATA/변수/외부데이터/환율데이터.csv')
df.drop('Unnamed: 0', axis=1, inplace=True)

df['REG_DATE'] = pd.to_datetime(df['REG_DATE'])
df['날짜'] = pd.to_datetime(df['날짜'])

price_list = df['REG_DATE'].unique()
price_list2 = df['날짜'].unique()


list_price = []
for i in df['환율']:
    list_price.append(i.replace(",", ""))


df['환율'] = list_price
# print(df)

dict_price = {}

for i in price_list:
    is_price = df['날짜'] == i
    price_df = df[is_price]
    # print(price_df)
    dict_price[i] = price_df['환율'].sum()

# print(df['환율'])
# print(dict_price)
dict_price = pd.DataFrame(dict_price, index=['환율'])


dict_price = dict_price.transpose()

dict_price['환율'].interpolate()

print(dict_price)

dict_price.to_csv('RAW_DATA/변수/외부데이터/환율데이터전처리.csv')
```

## 두바이유

```python
df = pd.read_csv('RAW_DATA/변수/외부데이터/두바이유내역(investing.com).csv')
df1 = pd.read_csv('RAW_DATA/변수/외부데이터/환율데이터.csv')

# print(df.columns)

df.drop('오픈', axis=1, inplace=True)
df.drop('고가', axis=1, inplace=True)
df.drop('저가', axis=1, inplace=True)
df.drop('거래량', axis=1, inplace=True)
df.drop('변동 %', axis=1, inplace=True)

date = []
for i in df['날짜']:
    i = i.replace("년", " -")
    i = i.replace("월", " -")
    i = i.replace("일", "")
    date.append(i)

# print(date)

df['날짜'] = date

df['날짜'] = pd.to_datetime(df['날짜'])
df1['REG_DATE'] = pd.to_datetime(df1['REG_DATE'])


# print(df)

do = {}
for i in df1['REG_DATE']:
    is_date = df['날짜'] == i
    date = df[is_date]
    do[i] = date['종가'].sum()

# print(do)

do_price = pd.DataFrame(do, index=['종가'])
do_price = do_price.transpose()
# print(do_price)

do_price.to_csv('RAW_DATA/변수/외부데이터/두바이유전처리.csv')
```
