# 연어 

```python
salmon = pd.read_csv('data/result/초안/salmon_result_initial.csv', encoding='cp949')

salmon.head(2)
```

```python
salmon_cloumns = []

for i in salmon.columns:
    salmon_cloumns.append(i)

print(salmon_cloumns)
len(salmon.columns)
```

```python
for i in range(1, len(salmon_cloumns) -1):
    value = salmon_cloumns[i]
    x_salmon = salmon[value]
    y_salmon = salmon['P_PRICE']
    corr = np.corrcoef(x_salmon, y_salmon)[0, 1]
    print(f'{value}과 P_PRICE의 상관계수 : {corr}')
```
